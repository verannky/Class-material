## Correlation

untuk menghitung correlation, kita bisa menggunakan fungsi corr yang disediakan oleh library pandas atau fungsi pearson/spearman untuk library scipy 

### corr

corr merupakan library yang disediakan oleh pandas untuk menghitung korelasi antara dua variabel numerik menggunakan berbagai macam correlation formula. Detail parameter yang ditangkap oleh fungsi ini adalah sebagai berikut:

Screen Shot 2021-10-03 at 13.15.48.png

Fungsi ini akan mengeluarkan matrix korelasi antara variabel-variabel numerik tersebut, dimana diagonalnya selalu bernilai 1

```python
## contoh corr 
df_titanic.corr(method='pearson')
```

## pearson

pearson merupakan function yang akan mengembalikan besar korelasi antara dua variabel yang diinputkan menggunakan formula pearson yang disediakan oleh library scipy. detail fungsi ini dapat dilihat di gambar berikut:

Screen Shot 2021-10-03 at 13.12.46.png

Output dari fungsi pearsonr adalah dua nilai: correlation value dan pvalue. Correlation value adalah besar korelasi antara dua variabel, sementara pvalue menjelaskan probability bahwa korelasi antara dua variabel tersebut adalah kebetulan (TBD in adv. statistic session)

Screen Shot 2021-10-03 at 13.27.14.png

```python
## contoh pearson
from scipy import stats
stats.pearsonr(df_titanic['Survived'], df_titanic['Fare'])
```

## spearman

Selain itu, library scipy juga menyediakan fungsi spearmanr untuk menghitung korelasi antara dua numerical data menggunakan rumus spearman. Fungsi ini mengambil lebih banyak argumen opsional dibandingkan pearsonr, disebabkan oleh approach perhitungan korelasi yang lebih kompleks. Detail fungsi tersebut bisa dilihat di gambar berikut: 

Screen Shot 2021-10-03 at 13.09.50.png

Sama dengan function pearsonr, method ini akan menghasilkan dua nilai: correlation value dan p value

```python
### contoh spearmanr
stats.spearmanr(df_titanic['Survived'], df_titanic['Fare'])
```

## Statistical Plot

Untuk memvisualisasikan korelasi dan probability distribution, kita bisa menggunakan library yang disediakan oleh python, seperti pandas, matplotlib, seaborn, dst. Untuk keperluan statistical plotting, kita hanya akan menggunakan librray visualization dasar yang disediakan oleh pandas

### histogram

Fungsi hist pada library pandas akan menampilkan distribusi dari satu dataframe, ataupun spesifik kolom tertentu. 

Screen Shot 2021-10-03 at 13.52.07.png

Dari beberapa argumen yang disediakan, salah satu yang perlu diperhatikan adalah parameter **bins**, karena nilai ini akan menjelaskan seberapa detail histogram yang akan dihasilkan. Semakin tinggi nilai bins, maka semakin sedikit pula jumlah bar yang ada pada histogram

```python
## contoh histogram
df_titanic['Fare'].hist()
```

### density plot

untuk membuat density plot atas persebaran nilai suatu variabel, kita bisa menggunakan function hist yang disediakan oleh library pandas

Screen Shot 2021-10-03 at 14.26.54.png

```python
## contoh density plot
df_titanic['Fare'].plot.density()
```

### scatter plot

scatterplot merupakan fungsi yang akan mengambil nilai dua variabel numerik dan membuat peta persebaran data sesuai dengan dua nilai yang dimasukkan. scatter plot sanagat berguna untuk mengecek korelasi antara dua variabel, ataupun untuk mendeteksi cluster/outlier berdasarkan variabel yang dipetakan.

Screen Shot 2021-10-03 at 13.58.41.png

```python
## contoh scatterplot
df_titanic.plot.scatter('Age','Fare')
```
