### Importing Text Files

Secara umum, langkah-langkah yang diperlukan untuk membaca text file dalam bahasa python adalah membuka file tersebut dan kemudian memproses file baris per baris. Dalam implementasinya, ada dua cara yang bisa digunakan.

### menggunakan library standard

dengan menggunakan cara ini, python akan membaca file tersebut menggunakan fungsi **open()** dan menampung isi dari file tersebut ke dalam satu variabel. setelah itu, kita perlu menambahkan fungsi **close()** untuk menutup file tersebut. Penutupan file ini harus dilakukan supaya file tersebut bisa tetap utuh dan tidak corrupt.

```python
# membaca textfile secara manual
f=open('/content/contoh_txt.txt','r')
print(f.read())
```

```python
# mengecek status variabel f
f.closed
```

```python
# menutup file
f.close()
```

### menggunakan with

dengan menggunakan statement with, kita tidak perlu menuliskan statement **close()** saat membaca text file. statement with berguna untuk menjabarkan dalam baris kode mana saja file tersebut akan 'hidup' dan digunakan. 

```python
# membaca text menggunakan with
with open('/content/contoh_txt.txt','r') as g:
    print(g.read())
```

cara ini sangat direkomendasikan karena secara otomatis, file akan tertutup setelah statement with berakhir. sehingga file bisa lebih aman dan kapasitas memori lebih besar.

```python
# cek status file yg dibuka
g.closed
```

### Pickle

Pickle merupakan bentuk tipe data yang sering digunakan dalam python. pickle sendiri berbentuk binary file sehingga tidak bisa dibaca menggunakan editor. Pickle biasa digunakan untuk menyimpan data dimana konsistensi dan keutuhan data menjadi prioritas, seperti hasil modelling, informasi data angka, ds. Walaupun begitu, pickle hanya bisa diakses oleh bahasa pemrograman Python sehingga penggunaanya kurang dianjurkan untuk multi-programming language environment. 

pickle merupakan library yg terpisah dari pandas. sehingga untuk menggunakannya, diperlukan syntax import

```python
### import pickle
import pickle
```

#### write to pickle

untuk menyimpan suatu tipe data ke dalam pickle, syntax yang bisa digunakan adalah **dump**, dengan parameter sebagai berikut:

pickle_dump.PNG

```python
### contoh menyimpan ke dalam file pickle
inet_provider=['indihome','indosat','first media','biznet','my republic']
pickle.dump(inet_provider, open("provider.pkl", "wb"))
```

#### read pickle

untuk membaca file pickle, syntax yang digunakan adalah load dengan parameter berupa nama file pickle yang akan dibaca oleh python. sebagai berikut:

load_pickle.PNG

```python
### contoh membaca file pickle
pickle.load(open("provider.pkl", "rb"))
```

## Querying relational databases in Python

### Cursor Class

object cursor merupakan class dari library psycopg2 yang memungkinkan user untuk memasukkan perintah sql dari python dan menjalankannya menggunakan data yang tersimpan di dalam database. class ini hanya bisa dibuat **setelah** koneksi berhasil dibuat. Kelas cursor ini berisikan perintah-perintah yang bisa digunakan untuk eksekusi perintah SQL

```python
### membuat cursor
cur=conn.cursor()
```

### Fungsi Execute

Fungsi execute merupakan fungsi dari class cursor yang digunakan untuk mengeksekusi perintah SQL ke database server yang telah terhubung. Fungsi ini mengambil satu parameter, yaitu syntax SQL yang akan dijalankan.

execute.PNG

```python
### contoh memasukkan data dengan execute
cur.execute("""INSERT INTO batch_kma.backup_employee(employee_id, employee_name) values ('1111','BTS')""")
```

### Fungsi Fetchall

Fungsi fetchall merupakan fungsi yang digunakan untuk mengambil data dari database setelah proses eksekusi SQL berhasil dilakukan. Fungsi ini akan mengembalikan semua baris data yang dihasilkan oleh syntax SQL tersebut. Fungsi ini akan mengembalikan data dalam bentuk list of tuples

```python
### contoh mengambil data employee 
cur.execute("SELECT * from batch_kma.master_employee")
rows=cur.fetchall()
for r in rows:
    print(r)
```

## Customize and filter database using SQL

Dalam kasus memasukkan query SQL dengan parameter yang berbeda, kita bisa gunakan skema parameterized query yang digunakan oleh psycopg2. Contoh parameterized query yang bisa digunakan adalah seperti berikut:


parameterized_query.jpg

Perlu diperhatikan bahwa jika kita hanya menggunakan satu parameter saja, nilai parameter yang perlu dipassing harus mengikuti bentuk tuplenya.

```python
### contoh memanggil data employee dengan gender laki-laki
gender='Laki-laki'
cur.execute('SELECT * FROM batch_kma.master_employee where gender=%s', (gender,))
cur.fetchall()
```

```python
### contoh memanggil data employee dengan gender laki-laki dan umur diatas 23
gender='Laki-laki'
umur=23
cur.execute('SELECT * FROM batch_kma.master_employee where gender=%s and age>%s', (gender, umur))
cur.fetchall()
```
