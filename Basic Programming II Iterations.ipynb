### Perulangan

perulangan adalah instruksi khusus dalam bahasa pemrograman dan algoritma yang digunakan untuk mengulang beberapa perintah sesuai dengan jumlah yang telah ditentukan. tujuannya adalah untuk mempermudah pengerjaan program dan untuk mempersingkat instruksi program.

program dibawah merupakan contoh baris kode python untuk menampilkan tulisan 'hello world' sebanyak 10 kali secara manual. hal ini sangat tidak dianjurkan karena membuat kode tidak efisien, rawan kesalahan, dan sulit dimaintain

```python
### program menampilkan 'hello world' sebanyak 10 kali
```

kelemahan dari code diatas dapat diatasi menggunakan algoritma perulangan dengan flowchart tampak seperti gambar berikut

perulangan.png

dari gambar diatas, dapat dilihat bahwa ada beberapa perintah yang akan dijalankan secara berulang, yaitu <b>Print Hello World</b> dan <b> count=count+1</b>. Perintah-perintah yang diulang ini biasa disebut sebagai <b>iterasi</b> dan iterasi akan dilakukan sebanyak jumlah perulangan.

### Elemen dari Perulangan

secara umum, statement perulangan dalam bahasa pemorgraman harus memiliki 3 elemen ini, yaitu:
- inisialisasi: pembuatan variabel counter (variabel penghitung)
- conditional statement/logical expression: syarat untuk satu iterasi akan dijalankan
- update counter: statement untuk penambahan/pengurangan counter setelah satu iterasi selesai dijalankan

loop.png

### Tipe Iterasi

Dalam algoritma pengulangan, ada dua macam iterasi yang biasanya diimplementasikan, yaitu:
- indefinite: perulangan yang dilakukan tanpa adanya informasi sampai berapa kali perulangan dilakukan. Iterasi akan dijalankan selama satu kondisi terpenuhi
- definite: perulangan yang dilakukan selama kondisi masih terpenuhi dan dalam batas jumlah perulangan yang akan dilakukan
